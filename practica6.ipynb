```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Definir los parmetros de la distribucin gaussiana
media = 1
desviacion_estandar = 2

# Generar datos para graficar la distribucin
x = np.linspace(media - 4 * desviacion_estandar, media + 4 * desviacion_estandar, 100)
y = norm.pdf(x, media, desviacion_estandar)

# Graficar el histograma
plt.plot(x, y, label='Distribucin Gaussiana')
plt.xlabel('Valor')
plt.ylabel('Densidad de probabilidad')
plt.title('Histograma de una distribucin gaussiana con media 1 y desviacin estndar 2')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def rango(dataset):
    return max(dataset) - min(dataset)

def leninv(dataset, intervals):
    return rango(dataset) / intervals

def sepbin(dataset, intervals, interval_len):
    bins = [[] for _ in range(intervals)]
    for data_point in dataset:
        bin_index = int((data_point - min(dataset)) / interval_len)
        if bin_index == intervals:
            bin_index -= 1  # In case the data point is equal to the maximum value
        bins[bin_index].append(data_point)
    return bins

campana = np.random.normal(1, 2, 100)
campana = sorted(campana)
num_intervals = 10
interval_len = leninv(campana, num_intervals)

bins = sepbin(campana, num_intervals, interval_len)

# Calcular las frecuencias para cada intervalo
frecuencias = [len(bin_) for bin_ in bins]

# Calcular el centro de cada intervalo para graficar la lnea
interval_centers = [min(campana) + i * interval_len + interval_len / 2 for i in range(num_intervals)]

# Graficar la lnea
plt.plot(interval_centers, frecuencias, marker='o', linestyle='-')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.title('Histograma de una distribucin gaussiana con media 1 y desviacin estndar 2')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def rango(dataset):
    return max(dataset) - min(dataset)

def leninv(dataset, intervals):
    return rango(dataset) / intervals

def sepbin(dataset, intervals, interval_len):
    bins = [[] for _ in range(intervals)]
    for data_point in dataset:
        bin_index = int((data_point - min(dataset)) / interval_len)
        if bin_index == intervals:
            bin_index -= 1  # En caso de que el punto de datos sea igual al valor mximo
        bins[bin_index].append(data_point)
    return bins

campana = np.random.normal(1, 2, 100)
campana = sorted(campana)
num_intervals = 10
interval_len = leninv(campana, num_intervals)

bins = sepbin(campana, num_intervals, interval_len)

# Calcular las frecuencias para cada intervalo
frecuencias = [len(bin_) for bin_ in bins]

# Graficar el histograma
plt.hist(campana, bins=num_intervals, edgecolor='black', alpha=0.7)

# Graficar cada punto en su subparticin
for i in range(num_intervals):
    plt.plot([i] * len(bins[i]), bins[i], 'o')

plt.xlabel('Intervalo')
plt.ylabel('Valor')
plt.title('Histograma con puntos en subparticiones')
plt.grid(True)
plt.show()
```
