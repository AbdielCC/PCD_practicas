```python
import numpy as np
import matplotlib.pyplot as plt

# Función para calcular la distancia euclidiana entre dos puntos
def distancia(a, b):
  return np.sqrt(np.sum((a - b) ** 2))

# Generar puntos aleatorios
np.random.seed(42)
num_puntos = 100
puntos = np.random.rand(num_puntos, 2)

# Definir el número de clusters (k)
k = 3

centroides = puntos[np.random.choice(puntos.shape[0], k,replace=False)]
etiquetas = np.zeros(num_puntos)

def asignar_puntos(puntos, centroides):
  for i in range(len(puntos)):
    distancias = [distancia(puntos[i], c) for c in centroides]
    etiquetas[i] = np.argmin(distancias)

# actualizar la posición de los centroides
def actualizar_centroides(puntos, etiquetas, k):
  nuevos_centroides = np.zeros((k, puntos.shape[1]))
  for i in range(k):
    nuevos_centroides[i] = np.mean(puntos[etiquetas == i], axis=0)
  return nuevos_centroides

#
iteraciones = 10

for _ in range(iteraciones):
  asignar_puntos(puntos, centroides)
  nuevos_centroides = actualizar_centroides(puntos, etiquetas, k)
  if np.all(centroides == nuevos_centroides):
    break
  centroides = nuevos_centroides

colores = ['r','g','b','y','c','m']
for i in range(num_puntos):
  plt.scatter(puntos[i, 0], puntos[i, 1], color=colores[int(etiquetas[i])])

# Grafico e impresion
plt.scatter(centroides[:, 0], centroides[:, 1], marker='*',s=200,color='black', label='centroides')

plt.legend()
plt.title('Algoritmo K-medios')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.show()
```
