Introducción.

En esta práctica, realizaremos un análisis exploratorio de la famosa base de datos Iris. Esta base de datos es un conjunto de datos clásico en el campo de la ciencia de datos y el aprendizaje automático, ya que contiene mediciones de cuatro características diferentes de flores de tres especies de Iris: Iris setosa, Iris versicolor y Iris virginica. Las cuatro características medidas son la longitud del sépalo, el ancho del sépalo, la longitud del pétalo y el ancho del pétalo. El objetivo de esta práctica es investigar las relaciones entre estas características mediante la generación de gráficos de dispersión y el ajuste de modelos polinómicos a los datos. Este análisis nos ayudará a comprender mejor cómo estas características están correlacionadas y qué patrones podemos observar en los datos.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Cargar la base de datos Iris
iris = pd.read_csv("C:/Users/Rulig/OneDrive/Escritorio/apps/progra/python/PCD/iris.csv")

# Definir una función para ajustar un polinomio y graficar los datos
def ajustar_y_graficar(x, y, grado=3):
    n = len(y)
    m = grado + 1
    X = np.ones((n, m))
    for i in range(n):
        for j in range(m):
            X[i][j] = pow(x[i], j)
    
    xt = np.transpose(X)
    xtx_i = np.linalg.inv(np.dot(xt, X))
    B = np.dot(np.dot(xtx_i, xt), y)
    
    y2 = []
    for i in range(len(x)):
        y2.append(sum(B[j] * x[i]**j for j in range(m)))
    
    plt.scatter(x, y, color='orange')
    plt.plot(x, y2)
    plt.show()
    
    print(B)

# Combinaciones de variables
combinaciones = [
    ("sepal_length", "sepal_width"),
    ("sepal_length", "petal_length"),
    ("sepal_length", "petal_width"),
    ("sepal_width", "petal_length"),
    ("sepal_width", "petal_width"),
    ("petal_length", "petal_width")
]

# Ajustar y graficar para cada combinación
for (var_x, var_y) in combinaciones:
    x = iris[var_x]
    y = iris[var_y]
    print(f"Ajuste para {var_x} vs {var_y}")
    ajustar_y_graficar(x, y)
```

1. Sepal Length vs Sepal Width


La gráfica muestra una dispersión de puntos que representan la relación entre la longitud del sépalo (sepal length) y el ancho del sépalo (sepal width).
La curva ajustada indica una relación no lineal, pero la dispersión de los datos sugiere que la relación no es muy fuerte.
En general, los puntos están bastante dispersos, lo que indica que la longitud del sépalo no es un buen predictor del ancho del sépalo.

2. Sepal Length vs Petal Length.


Esta gráfica muestra la relación entre la longitud del sépalo (sepal length) y la longitud del pétalo (petal length).
La curva ajustada sugiere una relación polinómica positiva, donde a medida que aumenta la longitud del sépalo, también aumenta la longitud del pétalo.
Los datos están más alineados a lo largo de la curva ajustada, indicando una relación más fuerte en comparación con la primera gráfica.

3. Sepal Length vs Petal Width.


Aquí se muestra la relación entre la longitud del sépalo (sepal length) y el ancho del pétalo (petal width).
La curva ajustada también sugiere una relación positiva, similar a la anterior, pero los datos muestran más dispersión en comparación con la longitud del pétalo.
Aunque hay una tendencia creciente, la relación parece ser menos fuerte que con la longitud del pétalo.

4. Sepal Width vs Petal Length.

Esta gráfica muestra la relación entre el ancho del sépalo (sepal width) y la longitud del pétalo (petal length).

La relación entre estas dos variables no parece ser fuerte, indicando que el ancho del sépalo no es un buen predictor de la longitud del pétalo.

5. Sepal Width vs Petal Width.

Esta gráfica muestra la relación entre el ancho del sépalo (sepal width) y el ancho del pétalo (petal width).
La curva ajustada muestra una relación polinómica negativa con bastante dispersión.
Al igual que con la longitud del pétalo, el ancho del sépalo no parece ser un buen predictor del ancho del pétalo.

6. Petal Length vs Petal Width.


Esta gráfica muestra la relación entre la longitud del pétalo (petal length) y el ancho del pétalo (petal width).
La curva ajustada muestra una clara relación positiva polinómica, indicando que a medida que aumenta la longitud del pétalo, también aumenta el ancho del pétalo.
Los puntos están bien alineados a lo largo de la curva ajustada, indicando una relación fuerte entre estas dos variables.
